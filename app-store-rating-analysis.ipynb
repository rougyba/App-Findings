```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df1 = pd.read_csv('AppleStore.csv')
df2 = pd.read_csv('appleStore_description.csv')
```

```python
df1.head(40)
df1.isnull().sum
df1.isnull().sum()
df1.duplicated().sum()
df1.dtypes
```

```python
df2.head(40)
df2.isnull().sum()
df2.duplicated().sum()
df2.dtypes
```

```python
df_subset = df1[['track_name', 'price', 'user_rating', 'prime_genre']]
```

```python
df_subset['price'].mean()
```

```python
df_subset['price'].min(), df_subset['price'].max()
```

```python
df_subset[df_subset['price'] == 0].shape[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_subset['price'], df_subset['user_rating'])
plt.xlabel('Price')
plt.ylabel('User Rating')
plt.title('Price vs. User Rating')
plt.show
```

```python
df_subset['prime_genre'].value_counts().head(1)
```

```python
avg_rating_per_genre = df_subset.groupby('prime_genre')['user_rating'].mean().sort_values(ascending=False)
avg_rating_per_genre.plot(kind='barh', color='skyblue')
plt.title('Average App Rating by Genre')
plt.xlabel('App Genre')
plt.ylabel('Average User Rating')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()


```
